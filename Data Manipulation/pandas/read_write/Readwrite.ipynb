to skip the first row then or the that has no valuess or missing value then use skiprows

or you want the remove the first row or header file then use "header'

```python
df = pd.read_excel("stock_data.xlsx", skiprows=1)
df
```

```python
df = pd.read_excel("stock_data.xlsx", header=1)
df
```

we can also remove the header and its by default insert the no . if we use the arugment header = None

```python
df = pd.read_excel("stock_data.xlsx", header=None)
df
```

we can use header=none and the name arugments then we can write the missing header in the dataset 

```python
df = pd.read_excel("stock_data.xlsx", header= None, names=["ticket","eps","revenue","price","people"])
df
```

if your csv file or dataset is very big and we want to see just few rows then we can use the nrows argument , 
it print the no. of rows we want excluding the header section

```python
df = pd.read_csv("stock_data.csv", nrows=3)
df
```

to cleaning up the not available ,n.a. values by then we can use the na_values argument, after using this 
all the n.a value replace with the NaN

```python
df = pd.read_csv("stock_data.csv", na_values= ["not available","n.a."])
df
```

we can also use the dist system to replace the not available value ,but we use the dist system because if we replace the some negative values 
but there is some columns where the where we want the negative value then in this case we need the dist system,

na_values use to convert it into the NaN

```python
df = pd.read_csv("stock_data.csv", na_values={
    'eps' : ["not available","n.a."],
    'revenue' : ["not available","n.a.",-1],
    'people' : ["not available","n.a."]
})
df
```

Writing the Csv 

```python
df.to_csv('new.csv',index=False)#index = false , it not give the index in the csv file 
```
